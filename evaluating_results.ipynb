```python
import pandas as pd
import numpy as np

from sklearn.metrics import brier_score_loss
from sklearn.calibration import calibration_curve, CalibrationDisplay
```

```python
file = r"C:\Users\yaobv\Downloads\tennis logs - Copy of Jon sheet.csv"
```

```python
df = pd.read_csv(file)

df = df.dropna().copy()
df.columns = ['p1_model_prob', 'p1', 'p2', 'p1_line', 'p1_market_prob', 'p2_line',
              'p2_market_prob', 'p1_result']

df['p1_model_prob'] = (df['p1_model_prob'] * 0.01).round(3)

df['model_loss'] = df['p1_model_prob'].sub(df['p1_result']) ** 2
df['market_loss'] = df['p1_market_prob'].sub(df['p1_result']) ** 2

df['adj_market_prob'] = (
    df
    .apply(lambda x: x['p1_market_prob'] / ((x['p1_market_prob'] + x['p2_market_prob']) / 1),
           axis=1)
)

df['model_market_diff'] = (df['p1_model_prob'] - df['p1_market_prob']).round(3)
df['model_market_bins'] = pd.cut(df['model_market_diff'],
                                 bins=[-0.3, -0.2, -0.1, 0, 0.1, 0.2, 0.3])

market_agree_cond = (
    ((df['p1_model_prob'] > 0.5) & (df['p1_market_prob'] > 0.5)) |
    ((df['p1_model_prob'] < 0.5) & (df['p1_market_prob'] < 0.5))
)

df['winner_agreement'] = np.where(market_agree_cond, 1, 0)

df.loc[df['p1_model_prob'] == 0.005, 'p1_model_prob'] = 0.5

print('model:', brier_score_loss(df['p1_result'], df['p1_model_prob']))
print('market:', brier_score_loss(df['p1_result'], df['p1_market_prob']))
```

```python
df.loc[df['winner_agreement'] == 0][['model_loss', 'market_loss']].head(200).expanding().mean().plot()
```

```python
model_true, model_pred = calibration_curve(df['p1_result'], df['p1_model_prob'], n_bins=10)
model_disp = CalibrationDisplay(model_true, model_pred, df['p1_model_prob'])

model_disp.plot()
```

```python
market_true, market_pred = calibration_curve(df['p1_result'], df['p1_market_prob'], n_bins=10)
market_disp = CalibrationDisplay(market_true, market_pred, df['p1_market_prob'])

market_disp.plot()
```
